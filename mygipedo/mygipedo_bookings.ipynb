# MyGipedo bookings

```python
%matplotlib inline
import pandas as pd
import numpy as np
import json
import seaborn as sns

```

```python
df = pd.read_json('mygipedo_bookings.json', lines=True)
```

```python
def expand(x):
    return x[0], x[1], pd.io.json.json_normalize(x[2])
```

```python
_df = df[['date', 'pitchId', 'bookings']].apply(expand, axis=1)
```

```python
__df = []

for row in _df:
    scalars = pd.DataFrame({'date': list(row[0].values())[0], 'pitchId': list(row[1].values())[0]}, index=[0])
    merged = pd.concat([scalars, row[2]], axis=1)
    __df.append(merged)
    
```

```python
bdf = pd.concat(__df, ignore_index=True)
```

```python
bdf = bdf.loc[(bdf.date > '2017-08-17')]
```

```python
bdf
```

```python
from dateutil import parser

bdf['DayOfWeek'] = bdf['date'].apply(lambda x: parser.parse(x).strftime("%A"))
```

```python
timeSlots = sorted(bdf['timeSlot'].unique())
```

```python
daysOfWeek = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
sns.set_context('poster')

g = sns.factorplot(x="timeSlot", data=bdf, kind="count",
                   palette="BuPu", size=6, aspect=1.5, order=timeSlots)
```

```python
j = sns.factorplot(x="DayOfWeek", data=bdf, kind="count",
                   palette="BuPu", size=6, aspect=1.5, order=daysOfWeek)
```

```python
users = bdf['userId.$oid'].unique()
users
```

```python
l = sns.factorplot(x="userId.$oid", data=bdf, kind="count",
                   palette="BuPu", size=6, aspect=1.5, order=users)
l.set_xticklabels(rotation=90)
```

```python
pitches = bdf['pitchId'].unique()
```

```python
for pitch in pitches:
    pitchBookings = bdf.loc[bdf['pitchId'] == pitch]
    f = sns.factorplot(x="timeSlot", data=pitchBookings, kind="count",
                   palette="BuPu", size=6, aspect=1.5, order=timeSlots)
    f.set(title=pitch)
```
